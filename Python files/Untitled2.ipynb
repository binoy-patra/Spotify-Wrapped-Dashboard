```python
import pandas as pd

# Load the main dataset
df_main = pd.read_csv("spotify_history.csv")
```

```python
import pandas as pd

# Assuming df_main is already available and 'ts' is in datetime format.
# Convert 'ts' to datetime if needed
df_main['ts'] = pd.to_datetime(df_main['ts'], errors='coerce')

# Extract year and month from 'ts'
df_main['year'] = df_main['ts'].dt.year
df_main['month'] = df_main['ts'].dt.month

# Group by 'year', 'month', 'track_name', then sum 'ms_played'
df_grouped = df_main.groupby(['year', 'month', 'track_name', 'spotify_track_uri'])['ms_played'].sum().reset_index()

# For each year and month, get the top 5 tracks based on the sum of 'ms_played'
df_top_tracks = df_grouped.groupby(['year', 'month']).apply(lambda x: x.nlargest(5, 'ms_played')).reset_index(drop=True)

# Show the result
df_top_tracks
```

```python
df_top_tracks.to_csv('Top Songs.csv', index= False)
```
